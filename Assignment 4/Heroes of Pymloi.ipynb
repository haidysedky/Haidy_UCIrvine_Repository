```python
import pandas as pd
```

```python
file = "heroes_pymoli.csv"
df = pd.read_csv(file)
df.head()
```

```python
#PLAYER COUNT
Total_Players_Number = df["SN"].nunique()
Total_Players_Number
pd.DataFrame({"Total Players":[Total_Players_Number]})
```

```python
#PURCHASING ANALYSIS (TOTAL)
Items_Numbers = df["Item ID"].nunique()
Avg_Price = "${:.2f}".format(df["Price"].mean())
Purchases_Number = df["Purchase ID"].count()
Total_Revenue = "${:.2f}".format(df["Price"].sum())
Purchasing_Analysis_df = pd.DataFrame({"Number of Unique Items":[Items_Numbers],
                                       "Average Price":[Avg_Price],
                                       "Number of Purchases":[Purchases_Number],
                                       "Total Revenue":[Total_Revenue]})
Purchasing_Analysis_df
```

```python
#GENDER DEMOGRAPHICS
df_grouped_by = df.groupby("Gender")
df_gender = df_grouped_by["SN"].nunique()
df_gender = pd.DataFrame(df_gender)
df_gender = df_gender.rename(columns={"SN":"Total Count"})
df_gender["Percentage of Players"] = ((df_gender["Total Count"]*100)/(Total_Players_Number)).map("{:.2f}%".format)
df_gender
```

```python
#GENDER PURCHASE ANALYSIS
df_gender_purchase = df_grouped_by["Purchase ID"].count()
df_gender_purchase = pd.DataFrame(df_gender_purchase)
df_gender_purchase = df_gender_purchase.rename(columns={"Purchase ID":"Purchase Count"})
df_gender_purchase["Average Purchase Price"] = (df_grouped_by["Price"].mean()).map("${:.2f}".format)
df_gender_purchase["Total Purchase Value"] = (df_grouped_by["Price"].sum()).map("${:.2f}".format)
df_gender_purchase
```

```python
#GENDER PURCHASE ANALYSIS CONTINUED
df_person_purchase = df.groupby(["Gender","SN"])
df_person_purchase = pd.DataFrame(df_person_purchase["Price"].sum())
df_person_purchase = df_person_purchase.groupby("Gender")
df_person_purchase = (df_person_purchase["Price"].mean()).map("${:.2f}".format)
df_gender_purchase["Avg Purchase Total per Person"] = df_person_purchase
df_gender_purchase
```

```python
#AGE DEMOGRAPHICS
bins = [0,9.9,14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99999]
age_labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
df["Age Categories"] = pd.cut(df["Age"], bins, labels=age_labels)
df_age = df.groupby("Age Categories")
df_age = pd.DataFrame(df_age["SN"].nunique()) 
df_age = df_age.rename(columns={"SN":"Count"})
df_age["Percentage"] = (df_age["Count"]*100/Total_Players_Number).map("{:.2f}%".format)
df_age
```

```python
#AGE PURCHASING ANALYSIS
df_age_group = df.groupby("Age Categories")
df_age_purchases = pd.DataFrame(df_age_group["Purchase ID"].count())
df_age_purchases = df_age_purchases.rename(columns={"Puchase ID":"Purchase Count"})
df_age_purchases["Average Purchase Price"] = (df_age_group["Price"].mean()).map("${:.2f}".format)
df_age_purchases["Total Purchase Value"] = (df_age_group["Price"].sum()).map("${:.2f}".format)
df_age_purchases
```

```python
#AGE PURCHASING ANALYSIS CONTINUED
df_by_person_group = df.groupby(["Age Categories","SN"])
df_by_person = pd.DataFrame(df_by_person_group["Price"].sum())
df_by_person = df_by_person.groupby("Age Categories")
df_age_purchases["Avergae Total Purchase per Person"] = (df_by_person["Price"].mean()).map("${:.2f}".format)
df_age_purchases
```

```python
#TOP 5 SPENDERS
df_groupby_spenders = df.groupby("SN")
df_spenders = pd.DataFrame(((df_groupby_spenders["Price"].sum()).sort_values(ascending=False)).map("${:.2f}".format))
df_spenders = df_spenders.rename(columns={"Price":"Total Purchase Value"})
df_spenders["Purchase Count"]= df_groupby_spenders["Purchase ID"].count()
df_spenders["Average Purchase Price"] = (df_groupby_spenders["Price"].mean()).map("${:.2f}".format)

df_spenders.head(5)
```

```python
#TOP 5 MOST POPULAR
df_groupby_item = df.groupby(["Item ID","Item Name","Price"])
df_most_popular = pd.DataFrame((df_groupby_item["Purchase ID"].count()).sort_values(ascending=False))
df_most_popular = df_most_popular.rename(columns={"Purchase ID":"Purchase Count"})
df_most_popular["Total Purchase Value"] = (df_groupby_item["Price"].sum()).map("${:.2f}".format)
df_most_popular.head(5)
```

```python
#TOP 5 MOST PROFITABLE
df_most_profitable = pd.DataFrame((df_groupby_item["Price"].sum()).sort_values(ascending=False))
df_most_profitable = df_most_profitable.rename(columns={"Price":"Total Purchase Value"})
df_most_profitable["Purchase Count"] = df_groupby_item["Purchase ID"].count()
df_most_profitable["Total Purchase Value"]=df_most_profitable["Total Purchase Value"].map("${:.2f}".format)
df_most_profitable.head(5)
```

# Three Observations from the above Data Analysis are:
-  The game of "Oathbreaker, Last Hope of the Breaking Storm" is the most popular and most profitable
-  The most of revenue comes from Male gender.
-  The most of revenue comes from the age category between 20 and 24
-  The least revenue comes from the age category of over 40

```python
#Another way to go around the last 2 sections for most popular and most profitable is:
df_groupby_item = df.groupby(["Item ID", "Item Name", "Price"])
df_purchase_analysis = (pd.DataFrame(df_groupby_item["Purchase ID"].count())).rename(columns={"Purchase ID":"Purchase Count"})
df_purchase_analysis["Total Purchase Value"] = df_groupby_item["Price"].sum()
df_most_popular = df_purchase_analysis.sort_values(by="Purchase Count", ascending=False)
df_most_popular["Total Purchase Value"] = df_most_popular["Total Purchase Value"].map("${:.2f}".format)
df_most_popular.head(5)
```

```python
df_most_profitable = df_purchase_analysis.sort_values(by="Total Purchase Value", ascending=False)
df_most_profitable["Total Purchase Value"] = df_most_profitable["Total Purchase Value"].map("${:.2f}".format)
df_most_profitable.head(5)
```

```python
#if we need to format the item price as well; it 'd be as follows:
df_most_profitable.reset_index(inplace=True)
df_most_profitable.iloc[:,2] = (pd.to_numeric(df_most_profitable.iloc[:,2])).map("${:.2f}".format)
df_most_profitable.head(5)
```
