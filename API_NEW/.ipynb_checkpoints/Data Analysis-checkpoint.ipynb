```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
listings_df = pd.read_csv("Top25ListingsinJewelryByScore.csv", index_col=0)
```

```python
listings_df.head(2)
```

```python
listings_grouped = listings_df.groupby("Categories")
listings_grouped_df = pd.DataFrame(listings_grouped["Favorites Number"].sum())
listings_grouped_df["Average Price"] = listings_grouped["Price"].mean()
listings_grouped_df["Average Views"] = listings_grouped["Number of Views"].mean()
listings_grouped_df.reset_index(inplace=True)
```
